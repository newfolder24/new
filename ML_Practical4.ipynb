```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Step 1: Load the Data
# Load the Iris dataset
df = pd.read_csv(r"C:\Users\VEDARTH KHANDVE\Downloads\CL_1 Prt_DK\CL_1 Prt_DK\Datasets\Iris.csv")

# Display the first few rows to understand the structure of the dataset
print("First few rows of the dataset:")
print(df.head())

# Step 2: Preprocess the Data
# Drop the 'Id' and 'Species' columns as they are not required for clustering
X = df.drop(columns=['Id', 'Species'])

# Standardize the data to normalize the feature scales
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 3: Elbow Method to determine optimal number of clusters
# Calculate the sum of squared distances (inertia) for a range of cluster values
inertia = []
k_values = range(1, 11)  # Testing for 1 to 10 clusters

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)  # Inertia is the sum of squared distances to cluster centers

# Plot the Elbow Graph
plt.figure(figsize=(8, 5))
plt.plot(k_values, inertia, marker='o', linestyle='-')
plt.title('Elbow Method to Determine Optimal k')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia (Sum of Squared Distances)')
plt.show()

# Step 4: Apply K-Means with optimal number of clusters
# From the elbow graph, choose k where the "elbow" or bend is located
# Assuming elbow occurs at k = 3 (common choice for Iris dataset)
optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
kmeans.fit(X_scaled)

# Add the cluster labels to the original dataset
df['Cluster'] = kmeans.labels_

# Display the first few rows of the dataset with the cluster labels
print("Dataset with cluster labels:")
print(df.head())

# Step 5: Visualize the Clusters (2D visualization using the first two features)
plt.figure(figsize=(8, 5))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=kmeans.labels_, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=100, c='red', marker='X')  # Cluster centers
plt.title('K-Means Clustering of Iris Dataset')
plt.xlabel('Feature 1 (Standardized)')
plt.ylabel('Feature 2 (Standardized)')
plt.show()
```

```python
# Step 4 (continued): Calculate and display the centroids for each cluster
centroids = kmeans.cluster_centers_

# Since we standardized the data, these centroids are also in the standardized scale.
# Let's print them out:
print("Centroids of each cluster (in standardized scale):")
print(centroids)

# Optional: Convert centroids back to the original scale
centroids_original_scale = scaler.inverse_transform(centroids)
print("\nCentroids of each cluster (in original scale):")
print(centroids_original_scale)

# Visualize centroids on the scatter plot
plt.figure(figsize=(8, 5))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=kmeans.labels_, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], s=100, c='red', marker='X', label="Centroids")  # Plot centroids
plt.title('K-Means Clustering of Iris Dataset with Centroids')
plt.xlabel('Feature 1 (Standardized)')
plt.ylabel('Feature 2 (Standardized)')
plt.legend()
plt.show()
```
